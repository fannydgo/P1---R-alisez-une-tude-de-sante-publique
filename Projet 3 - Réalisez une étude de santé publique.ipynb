## 3) Disponibilité alimentaire en kcal puis en kg de protéines (pour chaque produit et chaque pays)

## Pourcentage de protéines de chaque produit (pour chaque pays)

```python
pourcentageProt = animVégé.copy()
pourcentageProt['Ratio (prot/kg)'] = (pourcentageProt['Disponibilité de protéines en quantité (g/personne/jour)'] / 1000*365) / pourcentageProt['Disponibilité alimentaire en quantité (kg/personne/an)']
pourcentageProt['Protéines (%)'] = pourcentageProt['Ratio (prot/kg)'] * 100
pourcentageProt
```

## 5) Aliments les plus caloriques

```python
calAlim = animVégé.loc[(~animVégé["Disponibilité alimentaire (Kcal)"].isin(["inf", "NaN",0]))&(~animVégé["Disponibilité alimentaire en quantité (kg)"].isin(["inf", "NaN",0]))]
calAlim['Ratio (kcal/kg)'] = calAlim['Disponibilité alimentaire (Kcal)'] / calAlim['Disponibilité alimentaire en quantité (kg)']
calAlim = calAlim.groupby('Produit').mean()
calAlim.drop(['Population', 'Code zone', 'Code Produit'], axis = 1, inplace = True)
calAlim = calAlim.sort_values(by = 'Ratio (kcal/kg)', ascending = False)
calAlim.head(20)

```

## 6) Disponibilité intérieure mondiale des végétaux (en Kcal)

```python
calAlim = calAlim.loc[:,['Ratio (kcal/kg)']]
protAlim = protAlim.loc[:,['Ratio (prot/kg)']]
ratios = protAlim.join(calAlim, how = 'outer')
dispoVégé = animVégé [(animVégé["Origine"] == "Végétaux")]
dispoVégé = dispoVégé.groupby(['Produit', 'Code Produit'],as_index=True).sum()
dispoVégé = pd.merge(dispoVégé, ratios, left_index = True, right_index = True, how='left')
dispoVégé.drop(['Population', 'Code zone'], axis = 1, inplace = True)
dispoVégé['Disponibilité intérieure mondiale (Kcal)'] = dispoVégé['Disponibilité intérieure']*1000000*dispoVégé['Ratio (kcal/kg)']
dispoVégé['Disponibilité intérieure mondiale de protéines (kg)'] = dispoVégé['Disponibilité intérieure']*1000000* dispoVégé['Ratio (prot/kg)']
dispoVégé
```

```python
dispoMondiale = dispoVégé['Disponibilité intérieure mondiale (Kcal)'].sum()
print(dispoMondiale)
```
